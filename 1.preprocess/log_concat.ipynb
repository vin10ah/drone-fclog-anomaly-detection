```python
## 수정 버전

import os
import glob
import pandas as pd

def message_concat(base_dir, msg_lst):
    del_lst = []  # 삭제할 메시지 리스트, 필요 시 수정
    max = 0

    output_dir = os.path.join(base_dir, "all_results")
    os.makedirs(output_dir, exist_ok=True)

    for msg in msg_lst:
        if msg in del_lst:
            continue

        nor_paths = sorted(glob.glob(f"{base_dir}/normal/*/*{msg}.csv"))
        ab_paths = sorted(glob.glob(f"{base_dir}/abnormal/*/*{msg}.csv"))

        if not nor_paths:
            print(f"[경고] 정상 파일 없음: {msg}")
        if not ab_paths:
            print(f"[경고] 비정상 파일 없음: {msg}")

        nor_dfs = []
        ab_dfs = []

        for csv in nor_paths:
            try:
                df = pd.read_csv(csv)
                df["label"] = 0
                nor_dfs.append(df)
            except Exception as e:
                print(f"❌ 정상 파일 읽기 실패: {csv} → {e}")

        for csv in ab_paths:
            try:
                df = pd.read_csv(csv)
                df["label"] = 1
                ab_dfs.append(df)
            except Exception as e:
                print(f"❌ 비정상 파일 읽기 실패: {csv} → {e}")

        if not nor_dfs:
            print(f"[알림] '{msg}'에 대한 정상 데이터프레임 없음.")
            combined_nor = pd.DataFrame()
        else:
            combined_nor = pd.concat(nor_dfs, ignore_index=True)

        if not ab_dfs:
            print(f"[알림] '{msg}'에 대한 비정상 데이터프레임 없음.")
            combined_ab = pd.DataFrame()
        else:
            combined_ab = pd.concat(ab_dfs, ignore_index=True)

        if combined_nor.empty and combined_ab.empty:
            print(f"⚠️ '{msg}'는 정상/비정상 모두 비어 있어서 저장하지 않음.")
            continue

        result_df = pd.concat([combined_nor, combined_ab], ignore_index=True)

        if len(result_df) > max:
            max = len(result_df)

        save_path = os.path.join(output_dir, f"{msg}_merged.csv")
        result_df.to_csv(save_path, index=False)
        print(f"✅ 저장 완료: {save_path}")
    
    print(max)
```

```python
import pandas as pd
import glob
import os

# 1. 경로 설정 및 파일 목록 수집
paths = sorted(glob.glob("../data/all_results/*.csv"))

# 2. 결과 저장 리스트
msg_info = []

# 3. 각 파일 처리
for path in paths:
    try:
        df = pd.read_csv(path)
        num_rows = len(df)

        if num_rows < 10000:
            msg_field = os.path.basename(path).replace("_merged.csv", "")
            msg_info.append({"msg_field": msg_field, "num_rows": num_rows})

    except Exception as e:
        print(f"❌ {path} 읽기 실패: {e}")

# 4. 결과를 데이터프레임으로 정리
result_df = pd.DataFrame(msg_info)

result_df
```
