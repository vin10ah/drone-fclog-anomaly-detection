```python
def message_concat(base_dir, msg_lst):
    del_lst = ["ARM","CAND","FILE","EV","EAHR","MULT","MODE","ORGN","RCO2","RCI2",
               "VER","UNIT","XKFS","FMTU","FMT","MSG","PARM","STAK"]
    
    # del_lst = ["FMT"]


    output_dir = os.path.join(base_dir, "results")

    # 결과 폴더가 없으면 생성
    os.makedirs(output_dir, exist_ok=True)

    for msg in msg_lst:
        if msg in del_lst:
            continue
        
        # 각 폴더별 메세지 필드 파일 경로 추출
        nor_paths = sorted(glob.glob(f"{base_dir}/normal/*/*{msg}.csv"))
        ab_paths = sorted(glob.glob(f"{base_dir}/abnormal/*/*{msg}.csv"))

        nor_dfs = []
        ab_dfs = []

        # 정상/이상 라벨 추가 & 데이터프레임 concat
        for csv in nor_paths:
            try:
                df = pd.read_csv(csv)
                df["label"] = 0
                nor_dfs.append(df)
            except Exception as e:
                print(f"Failed to read normal file {csv}: {e}")

        for csv in ab_paths:
            try:
                df = pd.read_csv(csv)
                df["label"] = 1
                ab_dfs.append(df)
            except Exception as e:
                print(f"Failed to read abnormal file {csv}: {e}")

        combined_nor = pd.concat(nor_dfs, ignore_index=True)
        combined_ab = pd. concat(ab_dfs, ignore_index=True)

        result_df = pd.concat([combined_nor, combined_ab], ignore_index=True)
        result_df.to_csv(os.path.join(output_dir, f"{msg}_merged.csv"), index=False)

        print(f"저장 완료: {output_dir}/{msg}_merged.csv")
```
