```python
import os
import glob
import pandas as pd

# ✅ CSV 파일들이 저장된 경로
data_dir = "../data/results"  # 여기에 본인의 경로 입력
csv_paths = sorted(glob.glob(os.path.join(data_dir, "*.csv")))

print(f"총 {len(csv_paths)}개의 파일을 확인합니다.\n")

for path in csv_paths:
    
    try:
        df = pd.read_csv(path)
        df = df.drop('label', axis=1)
        filename = os.path.basename(path)
        total_rows = len(df)
        duplicate_rows = df.duplicated().sum()

        if duplicate_rows > 0:
            print(f"⚠️  {filename}: 중복 {duplicate_rows}개 발견 (총 {total_rows}행 중)")
        else:
            print(f"✅ {filename}: 중복 없음")

    except Exception as e:
        print(f"❌ {os.path.basename(path)}: 오류 발생 - {e}")

print(max)
```

```python
tab_df = pd.read_csv("../7.selected_features/results_20250602_1차/tabnet/val_metrics.csv")
ml_df = pd.read_csv("../7.selected_features/results_20250602_1차/ml_shap/selected_fields_model_comparison.csv")
```

```python
tab_df["model"] = "TabNet"
n_tab_df = tab_df[["msg_name", "model", "recall", "precision", "f1", "accuracy", "roc_auc"]]
```

```python
# TabNet에 Model 이름 추가
tab_df['model'] = 'TabNet'

tab_df = tab_df.rename(columns={
    'msg_name': 'msg_name',
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1',
    'roc_auc': 'roc_auc'
})

tab_df = tab_df[['msg_name', 'model', 'accuracy', 'precision', 'recall', 'f1', 'roc_auc']]

# ml_df 컬럼 정리
ml_df = ml_df.rename(columns={
    'field': 'msg_name',
    'Model': 'model',
    'Accuracy': 'accuracy',
    'Prec.': 'precision',
    'Recall': 'recall',
    'F1': 'f1',
    'AUC': 'roc_auc'
})
ml_df = ml_df[['msg_name', 'model', 'accuracy', 'precision', 'recall', 'f1', 'roc_auc']]

print("tab_df.columns:", tab_df.columns.tolist())
print("ml_df.columns:", ml_df.columns.tolist())


# 두 DataFrame 이어붙이기
final_df = pd.concat([tab_df, ml_df], ignore_index=True)

# msg_name, model 기준으로 정렬
final_df = final_df.sort_values(by=["msg_name", "model"]).reset_index(drop=True)

# 저장
final_df.to_csv("../7.selected_features/results_20250602_1차/re_selected_model_performance.csv", index=False)
print("저장 완료")
```
